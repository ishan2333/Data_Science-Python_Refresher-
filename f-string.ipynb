**F-String**

In Python f-string formatted string literal

To create an f-string, you prefix the string literal with the letter f (or F) and enclose any variables or expressions within curly braces {}. Python evaluates these expressions at runtime and replaces them with their resulting values. 

```python
name = "Ishan"
```

```python
city = "A.Nagar"
```

```python
state = "Maharashtra"
```

```python
print(f"My name is {name} i leave in {city} and i have pround on my state \"{state} \" .")
```

Expression inside f-string

```python
print(f"345+234 = {354+234}")
```

We can also write expressions inside curly braces {}

**1. Using format()**

```python
name2 = "Alice"
age2 = 25
print("My name is {} and I am {} years old.".format(name2, age2))
```

Positional and Keyword Arguments

```python
print("{0} is learning {1}".format("Alice", "Python")) # Using positions
print("{name} is learning {language}".format(name="Alice", language="Python"))
```
