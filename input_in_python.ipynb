**Handling Different Data Types (Type Casting)**

By default, input() always returns the input as a string, even if the user enters a number. To perform mathematical operations or use the input as a number, you must explicitly convert it using type casting functions like int() or float(). 

```python
age = int(input("Please enter your age : "))
```

```python
print(age)
```

```python
age += 5
```

```python
print(age)
```
