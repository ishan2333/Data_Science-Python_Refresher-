**For Loop**

For loop is a control flow statement used in programming to execute a block of code repeatedly a specific number of times or to iterate over a sequence

```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

Using range()

```python
for i in range(3):
    print(i) # Output: 0, 1, 2
```

.

**While Loop**

Runs as long as a condition is True .

```python
count = 0
while count < 3:
    print(count)
    count += 1
```
