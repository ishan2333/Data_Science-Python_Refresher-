**Logical Operators**

Logical operators are symbols or keywords that combine or invert Boolean conditions (true/false) to control program flow and decision-making

Used to combine conditional statements.

```python
X= True
Y = False
```

```python
X and Y
```

```python
X or Y
```

```python
not X
```

```python
not Y
```

**Bitwise Operators**

Perform bit-level operations

```python
Operator	Description     	    Syntax
&         	Bitwise AND	            x & y
|	        Bitwise OR	            x | y
~	        Bitwise NOT	            ~x
^	        Bitwise XOR	            x ^ y
>>	        Bitwise right shift 	x>>
<<	        Bitwise left shift	    x<<
```

```python
a = True
b = False
```

```python
a & b
```

```python
a | b
```
