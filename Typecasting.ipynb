Typecasting :- Typecasating is the process of converting a value from one data type to another.

```python
str1 = "56"         #string holding Number
```

```python
num1 = 233          #Int
```

```python
num2  = 54.54         #Float
```

```python
str2 = "Ishan"         #String Holding Carectors
```

```python
result_1 = int(str1)
```

```python
result_2 = str(num1)
```

```python
result_3 = int(num2)
```

```python
result_4 = list(str2)
```
