<h3>Json Module in Python</h3>

<b>Converting Python Objects to JSON (Serialization)</b>

```python
import json         #importing json module
```

```python
dict = {"Name" :"Ishan", "City" : "Nagar" , "age" :21}   #defining dictionary
type(dict)
```

```python
j = json.dumps(dict)       # converting python(dictionary) object to json
```

```python
print(j)               #printing the converted variable
type(j)               
```

```python
new_dict = json.loads(j)      
```

```python
print(new_dict)
type(new_dict)
```

<b>Converting  JSON to Python Objects(DeSerialization)</b>

```python
json_d = '{"Name": "Ishan", "City": "Nagar", "age": 21}'     #defining a json
```

```python
python_obj = json.loads(json_d)                      #convering json to python object
print(python_obj)
print(type(python_obj))
```
