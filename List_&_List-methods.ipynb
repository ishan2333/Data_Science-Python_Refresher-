**List and List Mehods**

A list in Python is an ordered, mutable collection of elements. It can contain elements of different types.

```python
my_list = []                # Empty List

numbers = [1, 2, 3, 4, 5]   # List with elements

mixed_list = [1, "Hello", 3.14, True]  # Mixed data types
```

```python
for num in numbers:
    print(num)
```

```python
for i in mixed_list:
    print(i)
```

**List Methods**

List methods are the functions which performs a specific operation on list

```python
fruits = ["apple", "banana", "cherry"]
fruits.append("orange")
print(fruits) # ['apple', 'banana', 'cherry', 'orange']
fruits.sort()
print(fruits) # ['apple', 'banana', 'cherry', 'orange']
```
