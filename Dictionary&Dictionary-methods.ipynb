**Dictionary & Dictionary Methods**

A dictionary in Python is an unordered, mutable, and key-value pair collection. It allows efficient data retrieval and modification. Dictionaries in Python are ordered as of Python 3.

```python
# Empty dictionary
empty_dict = {}
# Dictionary with key-value pairs
student = {
"name": "Alice",
"age": 25,
"grade": "A"
}
# Using dict() constructor
person = dict(name="John", age=30, city="New York")
```

**Accessing Dictionary Elements**

```python
# Using keys
print(student["name"]) # Alice
# Using get() (avoids KeyError if key doesn't exist)
print(student.get("age")) # 25
print(student.get("height", "Not Found")) # Default value
```
