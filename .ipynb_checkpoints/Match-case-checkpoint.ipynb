**Match Case Statements**

The match-case statement, introduced in Python 3.10, provides pattern similar to switch statements in other languages.

```python
age = int(input("Please enter your age"))
match age:
    case 0 :
        print("Child Don't need Ticket")
    case 10 :
        print("Child Need Half Ticket")
    case 18:
        print("Birthday...........!")
    case 15:
        print("Need Ticket")
    case _ :
        print("Invalid Input")
```
