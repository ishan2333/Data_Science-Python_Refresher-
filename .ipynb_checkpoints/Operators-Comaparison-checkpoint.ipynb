**Comparison Operators**

Comparison operators are used to compare two values

```python
a= 10
b=5
```

```python
a==b             #Equal to
```

```python
a!=b            #Not Equal to
```

```python
a>b             #Greater than
```

```python
a<b             #Less than
```

```python
a>=b            #Greater than or Equal to 
```

```python
a<=b            #Less Than or Equal to 
```
