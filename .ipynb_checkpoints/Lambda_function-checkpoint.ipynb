<h3>Lambda function</h3>

A lambda function in Python is an anonymous, single-expression function defined using the lambda keyword. It is commonly used for short, throwaway functions where a full function definition is unnecessary.

```python
# lambda arguments: expression
```

lambda → Keyword to define a lambda function<br>
arguments → Input parameters (comma-separated)<br>
expression → The operation performed (must be a single expression, not multiple statements)

```python
#example :
square = lambda x: x ** 2
print(square(5))
```

```python
# using map
numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, numbers))
print(squared)
```

```python
# using filter
numbers = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens) # Output: [2, 4, 6]
```

```python
from functools import reduce
numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```

<br><b>Lambda with Multiple Arguments</b>

```python
add = lambda x, y: x + y
print(add(3, 7)) # Output: 10

maximum = lambda x, y: x if x > y else y
print(maximum(10, 5))
```

<br><b>Lambda in Sorting Functions</b>

```python
students = [("Alice", 85), ("Bob", 78), ("Charlie", 92)]
students.sort(key=lambda student: student[1]) # Sort by score
print(students)
```

<br><b>When to Use Lambda Functions?</b>

<b>Use Lambda Functions When:</b><br>
• The function is short and simple<br>
• Used temporarily inside another function (e.g., map , filter ).<br>
• Avoiding defining a full function with def .<br>
<b>Avoid Lambda Functions When:</b><br>
• The function is complex (use def for readability).<br>
• Multiple operations/statements are needed<br>
