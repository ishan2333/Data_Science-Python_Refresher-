**If else statements**

In Python, conditional statements ( if , elif , and else ) are used to control the flow of a program based on conditions. These are essential in data science forhandling different scenarios in data processing, decision-making, and logic execution.

**Basic if Statement**

The if statement allows you to execute a block of code only if a condition is "True"

```python
z = 10
if z > 5:
    print("z is greater than 5")
```

.

**if-else Statement**

The else block executes when the if condition is False

```python
Y = 3
if Y > 5:
    print("Y is greater than 5")
else:
    print("Y is not greater than 5")
```

.

**if-elif-else Statement**

When multiple conditions need to be checked sequentially, use elif (short for “else if”)

```python
x = 5
if x > 10:
    print("x is greater than 10")
elif x > 5:
    print("x is greater than 5 but not more than 10")
elif x == 5:
    print("x is exactly 5")
else:
    print("x is less than 5")
```
